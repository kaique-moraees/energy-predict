# EDA - Série Temporal de Preços de Energia
**Análise Exploratória de Dados - Principais Insights**

## 1. Setup e Configurações

```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime, timedelta
import os
import random
import warnings

# Seed
seed = 42
os.environ['PYTHONHASHSEED'] = str(seed)
random.seed(seed)
np.random.seed(seed)

# Configurações
sns.set(style="white")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = '12'
plt.rcParams['figure.figsize'] = [10, 5]
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
np.set_printoptions(suppress=True)
warnings.filterwarnings('ignore')
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# caminhos relativos ao notebook atual
path_raw = "../data/raw/"

df_price = pd.read_csv(path_raw + 'price_vars_daily.csv')
```

## 2. Preparação dos Dados

```python
# Converter data e criar features temporais
df_price['anomesdia'] = pd.to_datetime(df_price['anomesdia'])
df_price = df_price.sort_values('anomesdia').reset_index(drop=True)

df_price['ano'] = df_price['anomesdia'].dt.year
df_price['mes'] = df_price['anomesdia'].dt.month
df_price['dia_semana'] = df_price['anomesdia'].dt.dayofweek
df_price['fim_semana'] = df_price['dia_semana'].isin([5, 6]).astype(int)

print(f"Período: {df_price['anomesdia'].min().date()} a {df_price['anomesdia'].max().date()}")
print(f"Total de dias: {len(df_price)}")
df_price.head()
```

## 3. Estatísticas Descritivas

```python
numeric_cols = ['geracao_gas_fossil', 'geracao_carvao_duro', 'carga_total_real', 
                'preco_dia_seguinte', 'preco_real', 'temperatura', 'velocidade_vento']

df_price[numeric_cols].describe().T
```

## 4. Visualização das Séries Temporais

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle('Séries Temporais - Visão Geral', fontsize=16, fontweight='bold')

# Preço Real
axes[0, 0].plot(df_price['anomesdia'], df_price['preco_real'], linewidth=0.8, color='#2E86AB')
axes[0, 0].set_title('Preço Real (€/MWh)', fontweight='bold')
axes[0, 0].set_ylabel('€/MWh')
axes[0, 0].grid(True, alpha=0.3)

# Preço Dia Seguinte
axes[0, 1].plot(df_price['anomesdia'], df_price['preco_dia_seguinte'], linewidth=0.8, color='#A23B72')
axes[0, 1].set_title('Preço Dia Seguinte (€/MWh)', fontweight='bold')
axes[0, 1].set_ylabel('€/MWh')
axes[0, 1].grid(True, alpha=0.3)

# Geração Gás Fóssil
axes[1, 0].plot(df_price['anomesdia'], df_price['geracao_gas_fossil'], linewidth=0.8, color='#F18F01')
axes[1, 0].set_title('Geração - Gás Fóssil (MW)', fontweight='bold')
axes[1, 0].set_ylabel('MW')
axes[1, 0].grid(True, alpha=0.3)

# Geração Carvão Duro
axes[1, 1].plot(df_price['anomesdia'], df_price['geracao_carvao_duro'], linewidth=0.8, color='#6A994E')
axes[1, 1].set_title('Geração - Carvão Duro (MW)', fontweight='bold')
axes[1, 1].set_ylabel('MW')
axes[1, 1].grid(True, alpha=0.3)

# Carga Total
axes[2, 0].plot(df_price['anomesdia'], df_price['carga_total_real'], linewidth=0.8, color='#BC4749')
axes[2, 0].set_title('Carga Total Real (MW)', fontweight='bold')
axes[2, 0].set_ylabel('MW')
axes[2, 0].grid(True, alpha=0.3)

# Temperatura
axes[2, 1].plot(df_price['anomesdia'], df_price['temperatura'], linewidth=0.8, color='#D62828')
axes[2, 1].set_title('Temperatura (K)', fontweight='bold')
axes[2, 1].set_ylabel('Kelvin')
axes[2, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Distribuições

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
fig.suptitle('Distribuições das Principais Variáveis', fontsize=16, fontweight='bold')

variables = [
    ('preco_real', 'Preço Real', '#2E86AB'),
    ('geracao_gas_fossil', 'Geração Gás Fóssil', '#F18F01'),
    ('geracao_carvao_duro', 'Geração Carvão', '#6A994E'),
    ('carga_total_real', 'Carga Total', '#BC4749'),
    ('temperatura', 'Temperatura', '#D62828'),
    ('velocidade_vento', 'Vel. Vento', '#4ECDC4')
]

for idx, (col, title, color) in enumerate(variables):
    row = idx // 3
    col_idx = idx % 3
    
    axes[row, col_idx].hist(df_price[col], bins=40, alpha=0.7, color=color, edgecolor='black')
    axes[row, col_idx].axvline(df_price[col].mean(), color='red', linestyle='--', linewidth=2, label='Média')
    axes[row, col_idx].axvline(df_price[col].median(), color='green', linestyle='--', linewidth=2, label='Mediana')
    axes[row, col_idx].set_title(title, fontweight='bold')
    axes[row, col_idx].set_xlabel(col)
    axes[row, col_idx].legend()
    axes[row, col_idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Análise de Sazonalidade

```python
# Médias mensais
monthly_avg = df_price.groupby('mes')[['preco_real', 'carga_total_real', 
                                        'geracao_gas_fossil', 'temperatura']].mean()

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Padrões Sazonais - Médias Mensais', fontsize=16, fontweight='bold')

# Preço Real
axes[0, 0].plot(monthly_avg.index, monthly_avg['preco_real'], marker='o', 
                linewidth=2, markersize=8, color='#2E86AB')
axes[0, 0].set_title('Preço Real por Mês', fontweight='bold')
axes[0, 0].set_xlabel('Mês')
axes[0, 0].set_ylabel('€/MWh')
axes[0, 0].set_xticks(range(1, 13))
axes[0, 0].grid(True, alpha=0.3)

# Carga Total
axes[0, 1].plot(monthly_avg.index, monthly_avg['carga_total_real'], marker='o', 
                linewidth=2, markersize=8, color='#BC4749')
axes[0, 1].set_title('Carga Total por Mês', fontweight='bold')
axes[0, 1].set_xlabel('Mês')
axes[0, 1].set_ylabel('MW')
axes[0, 1].set_xticks(range(1, 13))
axes[0, 1].grid(True, alpha=0.3)

# Geração Gás Fóssil
axes[1, 0].plot(monthly_avg.index, monthly_avg['geracao_gas_fossil'], marker='o', 
                linewidth=2, markersize=8, color='#F18F01')
axes[1, 0].set_title('Geração Gás Fóssil por Mês', fontweight='bold')
axes[1, 0].set_xlabel('Mês')
axes[1, 0].set_ylabel('MW')
axes[1, 0].set_xticks(range(1, 13))
axes[1, 0].grid(True, alpha=0.3)

# Temperatura
axes[1, 1].plot(monthly_avg.index, monthly_avg['temperatura'], marker='o', 
                linewidth=2, markersize=8, color='#D62828')
axes[1, 1].set_title('Temperatura por Mês', fontweight='bold')
axes[1, 1].set_xlabel('Mês')
axes[1, 1].set_ylabel('Kelvin')
axes[1, 1].set_xticks(range(1, 13))
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Análise por dia da semana
weekday_avg = df_price.groupby('dia_semana')[['preco_real', 'carga_total_real']].mean()
weekday_names = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Padrões Semanais', fontsize=16, fontweight='bold')

axes[0].bar(range(7), weekday_avg['preco_real'], alpha=0.7, color='#2E86AB')
axes[0].set_title('Preço Real por Dia da Semana', fontweight='bold')
axes[0].set_xlabel('Dia da Semana')
axes[0].set_ylabel('€/MWh')
axes[0].set_xticks(range(7))
axes[0].set_xticklabels(weekday_names)
axes[0].grid(True, alpha=0.3, axis='y')

axes[1].bar(range(7), weekday_avg['carga_total_real'], alpha=0.7, color='#BC4749')
axes[1].set_title('Carga Total por Dia da Semana', fontweight='bold')
axes[1].set_xlabel('Dia da Semana')
axes[1].set_ylabel('MW')
axes[1].set_xticks(range(7))
axes[1].set_xticklabels(weekday_names)
axes[1].grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## 7. Decomposição Sazonal

```python
# Decomposição STL
ts_data = df_price.set_index('anomesdia')['preco_real']
decomposition = seasonal_decompose(ts_data, model='additive', period=365)

fig, axes = plt.subplots(4, 1, figsize=(16, 12))
fig.suptitle('Decomposição Sazonal - Preço Real', fontsize=16, fontweight='bold')

decomposition.observed.plot(ax=axes[0], color='#2E86AB', linewidth=0.8)
axes[0].set_ylabel('Observado')
axes[0].set_title('Série Original', fontweight='bold')
axes[0].grid(True, alpha=0.3)

decomposition.trend.plot(ax=axes[1], color='#F18F01', linewidth=2)
axes[1].set_ylabel('Tendência')
axes[1].set_title('Tendência', fontweight='bold')
axes[1].grid(True, alpha=0.3)

decomposition.seasonal.plot(ax=axes[2], color='#6A994E', linewidth=0.8)
axes[2].set_ylabel('Sazonalidade')
axes[2].set_title('Componente Sazonal', fontweight='bold')
axes[2].grid(True, alpha=0.3)

decomposition.resid.plot(ax=axes[3], color='#BC4749', linewidth=0.5)
axes[3].set_ylabel('Resíduos')
axes[3].set_title('Resíduos', fontweight='bold')
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 8. Matriz de Correlação

```python
correlation_matrix = df_price[numeric_cols].corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.3f', cmap='RdBu_r', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, ax=ax,
            vmin=-1, vmax=1)
ax.set_title('Matriz de Correlação', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Correlações com preço real
print("\nCorrelações com Preço Real:")
print(correlation_matrix['preco_real'].sort_values(ascending=False))
```

## 9. Análise: Preço Real vs Preço Dia Seguinte

```python
df_price['diferenca_preco'] = df_price['preco_real'] - df_price['preco_dia_seguinte']
df_price['diferenca_preco_abs'] = np.abs(df_price['diferenca_preco'])

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Análise: Preço Real vs Preço Dia Seguinte', fontsize=16, fontweight='bold')

# Scatter
axes[0].scatter(df_price['preco_dia_seguinte'], df_price['preco_real'], 
                alpha=0.3, s=10, color='#2E86AB')
axes[0].plot([df_price['preco_dia_seguinte'].min(), df_price['preco_dia_seguinte'].max()],
             [df_price['preco_dia_seguinte'].min(), df_price['preco_dia_seguinte'].max()],
             'r--', linewidth=2, label='y=x')
axes[0].set_xlabel('Preço Dia Seguinte (€/MWh)')
axes[0].set_ylabel('Preço Real (€/MWh)')
axes[0].set_title('Correlação', fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Distribuição da diferença
axes[1].hist(df_price['diferenca_preco'], bins=50, alpha=0.7, color='#9B59B6', edgecolor='black')
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[1].axvline(x=df_price['diferenca_preco'].mean(), color='green', linestyle='--', linewidth=2, 
                label=f'Média: {df_price["diferenca_preco"].mean():.2f}')
axes[1].set_xlabel('Diferença (€/MWh)')
axes[1].set_ylabel('Frequência')
axes[1].set_title('Distribuição das Diferenças', fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Série temporal da diferença
axes[2].plot(df_price['anomesdia'], df_price['diferenca_preco'], 
             linewidth=0.5, alpha=0.6, color='#9B59B6')
axes[2].axhline(y=0, color='red', linestyle='--', linewidth=1)
axes[2].set_ylabel('Diferença (€/MWh)')
axes[2].set_title('Diferença ao Longo do Tempo', fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Métricas
print("\nMétricas de Acurácia da Previsão:")
print(f"MAE: €{df_price['diferenca_preco_abs'].mean():.2f}/MWh")
print(f"RMSE: €{np.sqrt((df_price['diferenca_preco']**2).mean()):.2f}/MWh")
print(f"Correlação: {df_price[['preco_real', 'preco_dia_seguinte']].corr().iloc[0,1]:.4f}")
```

## 10. Resumo e Insights

```python
print("="*80)
print("RESUMO FINAL E PRINCIPAIS INSIGHTS")
print("="*80)

print(f"\nPERÍODO: {df_price['anomesdia'].min().date()} a {df_price['anomesdia'].max().date()}")
print(f"Total de observações: {len(df_price):,} dias")

print(f"\nPREÇO REAL:")
print(f"Média: €{df_price['preco_real'].mean():.2f}/MWh")
print(f"Mediana: €{df_price['preco_real'].median():.2f}/MWh")
print(f"Min: €{df_price['preco_real'].min():.2f}/MWh | Max: €{df_price['preco_real'].max():.2f}/MWh")
print(f"Volatilidade (CV): {(df_price['preco_real'].std() / df_price['preco_real'].mean() * 100):.2f}%")

print(f"\nCARGA E GERAÇÃO:")
print(f"Carga Total Média: {df_price['carga_total_real'].mean():,.0f} MW")
print(f"Geração Gás Fóssil Média: {df_price['geracao_gas_fossil'].mean():,.0f} MW")
print(f"Geração Carvão Duro Média: {df_price['geracao_carvao_duro'].mean():,.0f} MW")

print(f"\nACURÁCIA DA PREVISÃO:")
print(f"MAE: €{df_price['diferenca_preco_abs'].mean():.2f}/MWh")
print(f"RMSE: €{np.sqrt((df_price['diferenca_preco']**2).mean()):.2f}/MWh")

seasonal_diff = df_price.groupby('mes')['preco_real'].mean()
print(f"\nSAZONALIDADE:")
print(f"Mês mais caro: {seasonal_diff.idxmax()} (€{seasonal_diff.max():.2f}/MWh)")
print(f"Mês mais barato: {seasonal_diff.idxmin()} (€{seasonal_diff.min():.2f}/MWh)")

weekday_diff = df_price.groupby('fim_semana')['preco_real'].mean()
print(f"\nPreço dias úteis: €{weekday_diff[0]:.2f}/MWh")
print(f"Preço fins de semana: €{weekday_diff[1]:.2f}/MWh")

print("\n" + "="*80)
print("EDA CONCLUÍDO!")
print("="*80)
```
