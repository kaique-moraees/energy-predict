```python
## Bibliotecas De Manipulação de Dados e Visualização
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import plotly.express as px
from IPython.display import display, Image

## Bibliotecas de Modelagem Matemática e Estatística
import numpy as np
import scipy as sp
from statsmodels.tsa.seasonal import seasonal_decompose

## Bibliotecas de Manipulação de Tempo
from datetime import datetime, timedelta

## Seed (Geração)
import os
import random

# ========== SEED GLOBAL ==========
seed = 42
os.environ['PYTHONHASHSEED'] = str(seed)
random.seed(seed)
np.random.seed(seed)

## Configurações de visualização
import warnings
%matplotlib inline
sns.set(style="white")
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = '12'
plt.rcParams['figure.figsize'] = [10, 5]
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', lambda x: '%.2f' % x) # Não usa o formato formato de Notação Científica em Pandas
np.set_printoptions(suppress=True) # Não usa o formato de Notação Científica em Numpy Arrays
warnings.filterwarnings('ignore')
warnings.simplefilter(action='ignore', category=FutureWarning) # Remove os avisos de Future Warnings
```

```python
# caminhos relativos ao notebook atual
path_raw = "../data/raw/"

df_energy = pd.read_csv(path_raw + 'energy_dataset.csv')
df_weather = pd.read_csv(path_raw + 'weather_features.csv')
```

```python
# Renomear a coluna
df_energy = df_energy.rename(columns={'price actual': 'preco_energia'})

df_energy['preco_energia'].describe()
```

```python
# Converter para datetime com fuso horário
df_energy['time'] = pd.to_datetime(df_energy['time'], utc=True)

# Criar colunas de hora, dia, mês e ano
df_energy['hora'] = df_energy['time'].dt.hour
df_energy['dia'] = df_energy['time'].dt.day
df_energy['mes'] = df_energy['time'].dt.month
df_energy['ano'] = df_energy['time'].dt.year

# Estatísticas do preço por ano
estatisticas_ano = df_energy.groupby('ano')['preco_energia'].describe()

estatisticas_ano
```

```python
df_monthly = df_energy.resample('M', on='time')['preco_energia'].agg(['mean', 'std', 'min', 'max'])
# Supondo que o DataFrame se chama df_monthly

# Garante que o índice é datetime
df_monthly.index = pd.to_datetime(df_monthly.index)

# Cria coluna com o ano
df_monthly['year'] = df_monthly.index.year

# Seleciona apenas os anos desejados
anos = [2015, 2016, 2017, 2018]
df_filtered = df_monthly[df_monthly['year'].isin(anos)]

# Cria a grade de subplots 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharey=False)
axes = axes.flatten()

# Loop para criar um gráfico por ano
for i, ano in enumerate(anos):
    ax1 = axes[i]
    ax2 = ax1.twinx()

    df_ano = df_filtered[df_filtered['year'] == ano]

    # Linha e faixa do preço médio
    ax1.plot(df_ano.index, df_ano['mean'], color='steelblue', linewidth=2, label='Preço médio')
    ax1.fill_between(df_ano.index, df_ano['min'], df_ano['max'],
                     color='skyblue', alpha=0.2, label='Faixa min–máx')

    # Linha do desvio padrão
    ax2.plot(df_ano.index, df_ano['std'], color='orange', linewidth=2, linestyle='--', label='Desvio padrão')

    # Títulos e eixos
    ax1.set_title(f'{ano}', fontsize=13)
    ax1.set_xlabel('')
    ax1.set_ylabel('Preço da Energia')

    # Grades e legendas
    ax1.grid(True, linestyle='--', alpha=0.5)
    lines_1, labels_1 = ax1.get_legend_handles_labels()
    lines_2, labels_2 = ax2.get_legend_handles_labels()
    ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left', fontsize=9)

# Remove subplots vazios (caso não use todos)
for j in range(len(anos), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# =====================================
# Distribuição do Preço de Energia
# =====================================

# Criação de figura com dois plots lado a lado
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# -------------------------
# Violin Plot + Boxplot
# -------------------------
sns.violinplot(y=df_energy['preco_energia'], ax=ax[0], color='skyblue', inner=None)
sns.boxplot(y=df_energy['preco_energia'], ax=ax[0], width=0.2, color='orange')
ax[0].set_ylabel('Preço da Energia')
ax[0].set_xlabel('')

# -------------------------
# Histograma
# -------------------------
sns.histplot(df_energy['preco_energia'], bins=30, kde=True, ax=ax[1], color='skyblue')
ax[1].set_xlabel('Preço da Energia')
ax[1].yaxis.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Filtra apenas os anos desejados
anos = [2015, 2016, 2017, 2018]
df_filtrado = df_energy[df_energy['ano'].isin(anos)]

# Figura maior: 16x8 para melhor visualização
fig, axes = plt.subplots(2, len(anos), figsize=(18, 10))

# Determina limites comuns de y
y_min = df_filtrado['preco_energia'].min()
y_max = df_filtrado['preco_energia'].max()

# Para padronizar histograma: mesmo range no eixo x
x_min = df_filtrado['preco_energia'].min()
x_max = df_filtrado['preco_energia'].max()

for i, ano in enumerate(anos):
    df_ano = df_filtrado[df_filtrado['ano'] == ano]

    # --- Linha 1: Violin + Box ---
    sns.violinplot(y=df_ano['preco_energia'], ax=axes[0, i],
                   color='skyblue', inner=None)
    sns.boxplot(y=df_ano['preco_energia'], ax=axes[0, i],
                width=0.15, color='orange')
    axes[0, i].set_ylim(y_min, y_max)
    axes[0, i].set_xlabel('')
    axes[0, i].set_ylabel('Preço' if i == 0 else '')
    axes[0, i].set_title(f'{ano}', fontsize=12)
    axes[0, i].grid(False)

    # --- Linha 2: Histograma ---
    sns.histplot(df_ano['preco_energia'], bins=25, kde=True,
                 ax=axes[1, i], color='skyblue')
    axes[1, i].set_xlim(x_min, x_max)
    axes[1, i].set_xlabel('Preço')
    axes[1, i].set_ylabel('Frequência' if i == 0 else '')
    if i != 0:
        axes[1, i].set_yticklabels([])  # remove valores do eixo y das outras colunas
    axes[1, i].grid(False)

plt.tight_layout(pad=2.0)
plt.show()
```

```python
# =====================================
# Série Temporal — Preço de Energia
# Concept Drift + Máx/Mín discretos e deslocados
# =====================================

# Série diária
df_daily = (
    df_energy
    .resample('D', on='time')['preco_energia']
    .mean()
    .reset_index()
)

# Rolling mean (7 dias)
df_daily['rolling_mean'] = df_daily['preco_energia'].rolling(window=7, center=True).mean()

# -------------------------
# 🔸 Detectar possíveis concept drifts
# -------------------------
df_daily['rolling_90'] = df_daily['preco_energia'].rolling(window=90).mean()
changes = df_daily['rolling_90'].diff().abs()
threshold = changes.std() * 3
drift_points = df_daily.loc[changes > threshold, ['time', 'rolling_mean']].dropna()

# -------------------------
# 🔹 Máximo e Mínimo globais
# -------------------------
max_idx = df_daily['preco_energia'].idxmax()
min_idx = df_daily['preco_energia'].idxmin()
max_point = df_daily.loc[max_idx]
min_point = df_daily.loc[min_idx]

# -------------------------
# 🎨 Figura
# -------------------------
plt.figure(figsize=(15, 6))
plt.style.use('default')  # fundo branco

# Linha base
sns.lineplot(data=df_daily, x='time', y='preco_energia',
             color='steelblue', linewidth=1.2, alpha=0.6, label='Preço diário')

# Média móvel
plt.plot(df_daily['time'], df_daily['rolling_mean'],
         color='orange', linewidth=2, label='Média móvel (7 dias)')

# Concept drift (pontos vermelhos)
plt.scatter(drift_points['time'], drift_points['rolling_mean'],
            color='red', s=60, marker='^', label='Possível mudança de conceito', zorder=5)

# Pontos de máximo e mínimo — bolinhas menores e deslocadas
highlight_color = 'dimgray'
offset = 0.02 * (df_daily['preco_energia'].max() - df_daily['preco_energia'].min())

plt.scatter([max_point['time'], min_point['time']],
            [max_point['preco_energia'], min_point['preco_energia']],
            color=highlight_color, s=40, marker='o', zorder=6)

# Texto ligeiramente acima do ponto
plt.text(max_point['time'], max_point['preco_energia'] + offset,
         f"Máx: {max_point['preco_energia']:.2f}", color=highlight_color,
         fontsize=9, fontweight='bold', ha='center', va='bottom')
plt.text(min_point['time'], min_point['preco_energia'] + offset,
         f"Mín: {min_point['preco_energia']:.2f}", color=highlight_color,
         fontsize=9, fontweight='bold', ha='center', va='bottom')

# -------------------------
# Ajustes visuais
# -------------------------
plt.ylabel('Preço da Energia')
plt.xlabel('')
plt.grid(False)  # sem grid de fundo
plt.legend()
plt.xlim(df_daily['time'].min(), df_daily['time'].max())

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

plt.tight_layout()
plt.show()
```

```python
# ==============================
# Decomposição da Série Temporal
# ==============================

# Executa decomposição (modelo aditivo, período anual aproximado)
result = seasonal_decompose(df_daily['preco_energia'].interpolate(), model='additive', period=365)

# Configura estilo
sns.set_style("whitegrid")
sns.set_context("talk")

# Criação da figura
fig, axes = plt.subplots(4, 1, figsize=(14, 10), sharex=True)

# Série original
axes[0].plot(df_daily['preco_energia'], color='steelblue', linewidth=1.8)
axes[0].set_title('Série Original', fontsize=13, weight='bold')
axes[0].set_ylabel('Preço')

# Tendência
axes[1].plot(result.trend, color='orange', linewidth=2)
axes[1].set_title('Tendência de Longo Prazo', fontsize=13, weight='bold')
axes[1].set_ylabel('Tendência')

# Sazonalidade
axes[2].plot(result.seasonal, color='seagreen', linewidth=2)
axes[2].set_title('Sazonalidade Anual', fontsize=13, weight='bold')
axes[2].set_ylabel('Sazonalidade')

# Resíduo
axes[3].plot(result.resid, color='grey', linewidth=1.2)
axes[3].axhline(0, color='black', linestyle='--', linewidth=1)
axes[3].set_title('Componente Irregular (Resíduo)', fontsize=13, weight='bold')
axes[3].set_ylabel('Resíduo')

# Ajustes gerais
plt.suptitle('📊 Decomposição da Série Temporal — Tendência, Sazonalidade e Ruído',
             fontsize=16, weight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Teste ADF
result = adfuller(df_daily['preco_energia'].dropna())

print("=" * 50)
print("TESTE DE ESTACIONARIEDADE (Augmented Dickey-Fuller)")
print("=" * 50)
print(f"Estatística ADF: {result[0]:.4f}")
print(f"P-valor: {result[1]:.4f}")
print(f"Valores críticos:")
for key, value in result[4].items():
    print(f"  {key}: {value:.4f}")

if result[1] > 0.05:
    print("\n✓ Série NÃO estacionária (p > 0.05) - confirma observação visual")
else:
    print("\n✗ Série estacionária (p ≤ 0.05)")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decomposição da série
decomposition = seasonal_decompose(
    df_daily.set_index('time')['preco_energia'],
    model='additive',
    period=365  # período anual
)

# Plotar
fig, axes = plt.subplots(4, 1, figsize=(15, 10))
fig.suptitle('Decomposição da Série Temporal - Preços de Energia', fontsize=14, fontweight='bold')

decomposition.observed.plot(ax=axes[0], color='steelblue')
axes[0].set_ylabel('Observado')
axes[0].grid(False)

decomposition.trend.plot(ax=axes[1], color='orange')
axes[1].set_ylabel('Tendência')
axes[1].grid(False)

decomposition.seasonal.plot(ax=axes[2], color='green')
axes[2].set_ylabel('Sazonalidade')
axes[2].grid(False)

decomposition.resid.plot(ax=axes[3], color='red', alpha=0.6)
axes[3].set_ylabel('Resíduo')
axes[3].set_xlabel('Data')
axes[3].grid(False)

plt.tight_layout()
plt.show()
```

```python
# Adicionar coluna de ano
df_daily['ano'] = df_daily['time'].dt.year

# Estatísticas por ano
stats_ano = df_daily.groupby('ano')['preco_energia'].agg([
    ('Média', 'mean'),
    ('Mediana', 'median'),
    ('Desvio Padrão', 'std'),
    ('Mínimo', 'min'),
    ('Máximo', 'max'),
    ('Amplitude', lambda x: x.max() - x.min())
])

# Coeficiente de variação
stats_ano['CV (%)'] = (stats_ano['Desvio Padrão'] / stats_ano['Média']) * 100

print("\n" + "=" * 80)
print("ESTATÍSTICAS DESCRITIVAS POR ANO")
print("=" * 80)
print(stats_ano.round(2))
print("\nCV (%) = Coeficiente de Variação indica volatilidade relativa")
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Análise de Autocorrelação - Preços de Energia', fontsize=14, fontweight='bold')

# ACF
plot_acf(df_daily['preco_energia'].dropna(), lags=50, ax=axes[0])
axes[0].set_title('Autocorrelação (ACF)')
axes[0].grid(False)

# PACF
plot_pacf(df_daily['preco_energia'].dropna(), lags=50, ax=axes[1])
axes[1].set_title('Autocorrelação Parcial (PACF)')
axes[1].grid(False)

plt.tight_layout()
plt.show()
```
